```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.style.use('seaborn-muted') #can't plot without my favorite theme
```

```python
x = np.linspace(-10, 10, 1000000)
y = (1/(3*np.sqrt(2*np.pi)))*np.exp(-0.5*(x/3)**2)
```

```python
plt.plot(x, y)
plt.xlabel('x [unitless]')
plt.ylabel('y [also unitless]')
plt.title('Gaussian Curve')
plt.show()
```

```python
np.trapz(y, x)
```
